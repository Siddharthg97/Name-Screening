```python
import xgboost
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from dalex import Explainer

# Load the Iris dataset
data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.2, random_state=42)

# Train an XGBoost model
model = xgboost.XGBClassifier(objective="multi:softmax", num_class=3)
model.fit(X_train, y_train)

# Convert X_test to a DataFrame for dalex
X_test_df = pd.DataFrame(X_test, columns=data.feature_names)

# Wrap the XGBoost model using dalex explainer
xgb_explainer = Explainer(model, X_train, y_train, label="XGBoost")

# Explain a single prediction (record level)
prediction_instance = X_test_df.iloc[0, :]  # Use iloc directly on the DataFrame
local_feature_importance = xgb_explainer.predict_parts(
    new_observation=prediction_instance,
    type="break_down"
)

# Display the local feature importance for the first prediction
print("Local Feature Importance:")
print(list(local_feature_importance.result.contribution[1:-1]))
```

```python
def feature_attribution_dalex(x_test_value_row_number):
  # Wrap the XGBoost model using dalex explainer
  xgb_explainer = Explainer(model, X_train, y_train, label="XGBoost")

  # Explain a single prediction (record level)
  prediction_instance = X_test_df.iloc[x_test_value_row_number, :]  # Use iloc directly on the DataFrame
  local_feature_importance = xgb_explainer.predict_parts(
      new_observation=prediction_instance,
      type="break_down"
  )

  # Display the local feature importance for the first prediction
  contri_value=list(local_feature_importance.result.contribution[1:-1])
  return contri_value
```

```python
contri_lst=[]
for i in range(len(X_test_df)):
  val_arr=feature_attribution_dalex(i)
  contri_lst.append(val_arr)
```
